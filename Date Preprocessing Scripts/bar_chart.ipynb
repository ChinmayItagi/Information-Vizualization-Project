```python
import pandas as pd
import calendar
import json 
from collections import defaultdict
import copy
```

```python
df = pd.read_csv(r'C:\Users\Chinmay Itagi\Documents\GitHub\Information-Vizualization-Project_1\Data\crime-in-los-angeles\Crime_Data_2010_2017.csv')
list(df)
```

```python
df['year'] = pd.DatetimeIndex(df['Date Occurred']).year
df = df[['year',"Victim Age",'DR Number']]
```

```python
list(df)
```

```python
def getage(x):
    if x<=15:
        return "0-15"
        
    elif x>=15 and x<25:
        return "15-25"
         
    elif x>25 and x<=60:
        return "25-60"
    else: 
        return "above 60"
    
df["Victim Age"] =df["Victim Age"].apply(lambda x: getage(x))
df_group_sex = df.groupby(['year','Victim Age'],as_index=False).count()
```

```python
df_group_sex
```

```python
dict_values= defaultdict(dict)
for each in df_group_sex.iterrows():
    dict_values[each[1]['year']][each[1]["Victim Age"]] = each[1]["DR Number"]


```

```python
dict_values_copy = copy.deepcopy(dict_values)
```

```python
json_file = {}
for i,(j,k) in enumerate(dict_values_copy.items()):
    json_file[i]={"state":j,'freq':k,'total':sum(k.values())}
    
    
    
```

```python
json_file
```

```python
with open(r'C:\Users\Chinmay Itagi\Documents\GitHub\Information-Vizualization-Project_1\Data\Chinmay_Data\dict_data.json', 'w') as outfile:
    json.dump(json_file, outfile)
```
